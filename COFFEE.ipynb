```python
import pandas as pd
import numpy as np
import seaborn as svs
from google.colab import files
files.upload()
```

```python
cof=pd.read_csv("index (1).csv")
cof
```

```python
cof.isnull().sum()
```

```python
cof.dropna(subset=["card"],axis=0, inplace=True)
cof
```

```python
cof.isnull().sum()
```

```python
print(cof.describe())
```

```python
import matplotlib.pyplot as plt
```

```python
daily_sales = cof.groupby('date')['money'].sum().reset_index()
plt.figure(figsize=(10,5))
svs.lineplot(data=daily_sales, x='date', y='money', marker='o')
plt.title('Total Sales Over Time')
plt.ylabel('Total Revenue')
plt.xlabel('Date')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
coffee_counts=cof['coffee_name'].value_counts().reset_index()
coffee_counts.columns=['coffee_name','count']
plt.figure(figsize=(8,5))
svs.barplot(data=coffee_counts,x='count',y='coffee_name',palette='viridis')
plt.title('Most Popular Coffee Types')
plt.xlabel('Number of Sales')
plt.ylabel('Coffee Name')
plt.tight_layout()
plt.show()
```

```python
coffee_revenue=cof.groupby('coffee_name')['money'].sum().sort_values(ascending=False).reset_index()
plt.figure(figsize=(8,5))
svs.barplot(data=coffee_revenue, x='money', y='coffee_name', palette='magma')
plt.title('Revenue by Coffee Type')
plt.xlabel('Revenue (Money)')
plt.ylabel('Coffee Name')
plt.tight_layout()
plt.show()
```

```python
cof['datetime'] = pd.to_datetime(cof['datetime'])
cof['hour']=cof['datetime'].dt.hour
hourly_sales=cof.groupby('hour')['money'].sum().reset_index()
plt.figure(figsize=(10,5))
svs.lineplot(data=hourly_sales, x='hour', y='money', marker='o', color='green')
plt.title('Revenue by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Revenue')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
customer_freq=cof['card'].value_counts().reset_index()
customer_freq.columns=['card_id', 'visits']
plt.figure(figsize=(10,5))
svs.histplot(customer_freq['visits'], bins=10, kde=False)
plt.title('Distribution of Customer Visits')
plt.xlabel('Number of Visits')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```

```python
repeat_customers=customer_freq[customer_freq['visits']>1]
print(f"\nRepeat Customers:{len(repeat_customers)}/{customer_freq.shape[0]}")
```

```python
print("\nSummary Metrics:")
total_revenue = cof['money'].sum()
total_orders = cof.shape[0]
unique_customers = cof['card'].nunique()
print(f"Total Revenue: {total_revenue:.2f}")
print(f"Total Orders: {total_orders}")
print(f"Unique Customers: {unique_customers}")
```

```python
cof["datetime"]=pd.to_datetime(cof["datetime"])
cof["time"]=cof["datetime"].dt.time
print(cof)
```

```python
cof['datetime'] = pd.to_datetime(cof['datetime'])
cof['time_only']=cof['datetime'].dt.time
plt.figure(figsize=(10,5))
svs.stripplot(x='coffee_name',y=cof['datetime'].dt.hour+cof['datetime'].dt.minute/60,data=cof,jitter=True,palette='Set2', hue='coffee_name', legend=False)
plt.title('Transaction Time (Hour) by Coffee Type')
plt.xlabel('Coffee Type')
plt.ylabel('Time of Day (Hours)')
plt.show()
```

```python
hourly_items=cof.groupby(["hour","coffee_name"])["card"].count().reset_index()
hourly_items=hourly_items.rename(columns={"card":"purchase_count"})
plt.figure(figsize=(14,6))
svs.lineplot(x="hour", y="purchase_count", hue="coffee_name", data=hourly_items)
plt.title("Most Purchased Coffee by Hour of Day")
plt.xlabel("Hour of Day")
plt.ylabel("Purchases")
plt.legend(title="Coffee Type", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cof["coffee_name"]=le.fit_transform(cof["coffee_name"])
cof["cash_type"]=le.fit_transform(cof["cash_type"])
cof["card"]=le.fit_transform(cof["card"])
cof["datetime"]=le.fit_transform(cof["datetime"])
cof
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,8))
correlation_matrix = cof[['datetime', 'cash_type', 'card', 'money', 'coffee_name', 'hour']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
corr=cof.drop(columns=['date', 'time', 'time_only']).corr()
plt.figure(figsize=(8,6))
svs.heatmap(corr,annot=True,fmt=".2f",cmap="coolwarm",vmin=-1,vmax=1)
plt.title("Correlation Matrix Heatmap (fixed)")
plt.show()
```
